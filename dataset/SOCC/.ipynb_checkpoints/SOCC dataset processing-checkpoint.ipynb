```python
goal = df[['comment_text', 'toxicity_level', 'toxicity_level:confidence']]
```

```python
goal
```

```python
goal.columns = ['text', 'label', 'confidence']
```

```python
goal
```

```python
goal = goal[['text', 'confidence', 'label']]
goal
```

```python
good = goal.loc[~goal['label'].astype(str).str.contains('\n')]
print('Good:')
display(good)

bad = goal.loc[goal['label'].astype(str).str.contains('\n')]
print('Bad:')
display(bad)
```

```python
import numpy as np

# Process bad before.
okay = pd.DataFrame(columns=bad.columns)

for row in bad.itertuples():
    label = row[3]
    confidence = row[2]
    labels = label.split('\n')
    confidences = confidence.split('\n')

    if (labels[0] == '1' and labels[1] == '2') or \
        (labels[0] == '2' and labels[1] == '1'):
        new_confidence = np.max([float(confidences[0]), float(confidences[1])])
        new_row = pd.DataFrame({
            'text': [row[1]],
            'confidence': [str(new_confidence)],
            'label': ['normal'],
        })
        okay = pd.concat([okay, new_row], ignore_index=True)

    elif (labels[0] == '3' and labels[1] == '4') or \
        (labels[0] == '4' and labels[1] == '3'):
        new_confidence = np.max([float(confidences[0]), float(confidences[1])])
        new_row = pd.DataFrame({
            'text': [row[1]],
            'confidence': [str(new_confidence)],
            'label': ['toxic'],
        })
        okay = pd.concat([okay, new_row], ignore_index=True)
    else: 

        for l, c in zip(labels, confidences):
            la = 'normal'
            if l == '3' or l == '4':
                la = 'toxic'
            
            new_row = pd.DataFrame({
                'text': [row[1]],
                'confidence': [c],
                'label': [la],
            })
            okay = pd.concat([okay, new_row], ignore_index=True)
        
```

```python
# Push all good ones into okay
new_good = good.copy()
new_good.loc[(new_good['label'] == '1') | (new_good['label'] == '2'), 'label'] = 'normal'
new_good.loc[(new_good['label'] == '3') | (new_good['label'] == '4'), 'label'] = 'toxic'
new_good
```
